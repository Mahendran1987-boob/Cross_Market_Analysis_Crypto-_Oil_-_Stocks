```python
import requests
import pandas as pd
import time

all_records = []

for i in range(1, 6):   # 1 to 5 (5 pages)
    url = f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&per_page=250&order=market_cap_desc&page={i}&sparkline=false"

    response = requests.get(url)

    print(f"Page {i} status:", response.status_code)

    if response.status_code == 200:
        data = response.json()
        print(f"Page {i} records:", len(data))
        all_records.extend(data)
    else:
        print("Error on page", i)

    time.sleep(1)   # Prevent rate limit

print("Total records:", len(all_records))
#print(data)
df = pd.DataFrame(all_records)
df

cc =  df[['id','symbol','name','current_price','market_cap','market_cap_rank','total_volume','circulating_supply','total_supply','ath','atl','last_updated']]
#cc = pd.DataFrame(cc)
cc = cc.rename(columns={'last_updated': 'Date'})
cc['Date'] = pd.to_datetime(cc['Date']).dt.strftime('%Y-%m-%d')

cc.head()
#cc = cc.rename(columns={'last_updated': 'Date'})
cc_copy = cc.copy()
cc_copy.to_csv("/content/drive/MyDrive/Cryptocurrencies.csv")
#print(df)   # Should show (750, number_of_columns)
df.head()         # View first 5 rows
```

```python
#Correct code
import pandas as pd
import requests

coin_ids = ['bitcoin', 'ethereum', 'tether']

all_prices = []

for coin in coin_ids:

    url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart"

    params = {
        "vs_currency": "usd",
        "days": "365"
    }

    response = requests.get(url, params=params)
    r = response.json()

    df = pd.DataFrame(r['prices'], columns=['timestamp','price_usd'])

    df['date'] = pd.to_datetime(df['timestamp'], unit='ms')
    df['coin_id'] = coin

    df = df[['coin_id','date','price_usd']]

    all_prices.append(df)

crypto_prices = pd.concat(all_prices, ignore_index=True)
crypto_prices["date"] = pd.to_datetime(crypto_prices["date"]).dt.date
crypto_prices
crypto_prices = pd.concat(all_prices, ignore_index=True)
crypto_prices["date"] = pd.to_datetime(crypto_prices["date"]).dt.date
crypto_prices
crypto_prices.head()
```

```python
import yfinance as yf
import pandas as pd

# Choose indices
tickers = ["^GSPC", "^IXIC", "^NSEI"]
start_date = "2020-01-01"
end_date = "2025-09-30"

# Download historical data
stocks_df = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")
stocks_df.head()
#stocks_df.tail()
```

```python
stocks_1 = stocks_df['^GSPC'].reset_index() # collect 3 ticker data similarly , add a new column ticker and then concat them.
# handle the  nulls
stocks_1["Ticker"] = "^GSPC"
stocks_1 = stocks_1.dropna(subset=['Close'])
stocks_1.head()
```

```python
stocks_2 = stocks_df['^IXIC'].reset_index() # collect 3 ticker data similarly , add a new column ticker and then concat them.
# handle the  nulls
stocks_2["Ticker"] = "^IXIC"
stocks_2 = stocks_2.dropna(subset=['Close'])
stocks_2.head()
```

```python
stocks_3 = stocks_df['^NSEI'].reset_index() # collect 3 ticker data similarly , add a new column ticker and then concat them.
# handle the  nulls
stocks_3['Ticker'] = "^NSEI"
stocks_3 = stocks_3.dropna(subset=['Close'])
stocks_3.head()
```

```python
all_stocks = pd.concat([stocks_1, stocks_2, stocks_3])
all_stocks['Date'] = pd.to_datetime(all_stocks['Date']).dt.strftime('%Y-%m-%d')
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd

from streamlit_option_menu import option_menu
import sqlite3
conn = sqlite3.connect("ds.db")
cursor = conn.cursor()
st.title("💰🛢️📈Cross-Market Analysis")

with st.sidebar:
    selected = option_menu("Main Menu", ["Home", 'Filters',"Queries","Crypto Analysis"],
        icons=['house', 'gear'], menu_icon="cast", default_index=1)

if selected == "Home":
    st.write("This is my home page")
    st.write("Oil Market (e.g., Crude Oil)")
    st.write("Influences the global economy. When oil prices rise, costs increase for industries, which can reduce company profits.")
    st.write("Stock Market (e.g., S&P 500, NIFTY 50)")
    st.write("Reflects economic health. Rising oil prices often negatively impact stocks, especially transport and manufacturing sectors.")
    st.write("Cryptocurrency Market (e.g., Bitcoin, Ethereum)")
    st.write("Highly volatile and sometimes behaves independently, but can follow stock market trends during major economic events.")

if selected == "Filters":
     st.header("Crypto Oil Stock Market Analytics")
     filtered_df = pd.read_sql("select AVG(c.price_usd) AS bitcoin_avg_price, AVG(o.price_usd) AS oil_avg_price, AVG(s.close) AS sp500_avg_close, AVG(n.close) AS nifty_avg_close FROM crypto_price c JOIN oil_prices o ON c.Date = o.Date JOIN stock_prices s ON c.Date = s.Date AND s.ticker='^GSPC' JOIN stock_prices n  ON c.Date = n.Date AND n.ticker='^NSEI' WHERE c.coin_id='bitcoin' ",con=conn)
     btc_avg = filtered_df["bitcoin_avg_price"][0]
     oil_avg = filtered_df["oil_avg_price"][0]
     sp500_avg = filtered_df["sp500_avg_close"][0]
     nifty_avg = filtered_df["nifty_avg_close"][0]
     st.subheader("Market Average Prices")
     col1,col2,col3,col4 = st.columns(4)
     col1.metric("Bitcoin Avg Price", round(btc_avg,2))
     col2.metric("Oil Avg Price", round(oil_avg,2))
     col3.metric("S&P500 Avg Close", round(sp500_avg,2))
     col4.metric("NIFTY Avg Close", round(nifty_avg,2))
     st.subheader("Daily Market Snapshot")
     filtered_df = pd.read_sql("SELECT c.date, c.price_usd AS bitcoin_price, o.price_usd AS oil_price, s.close AS sp500_close, n.close AS nifty_close FROM crypto_price c JOIN oil_prices o ON c.date = o.date JOIN stock_prices s ON c.date = s.date AND s.ticker='^GSPC' JOIN stock_prices n ON c.date = n.date AND n.ticker='^NSEI' WHERE c.coin_id='bitcoin' GROUP BY c.date ORDER BY c.date; ",con=conn)
     st.dataframe(filtered_df)

if selected == "Queries":
     option = st.selectbox(
     "Ccryptocurrencies",
    (
     "1. Find the top 3 cryptocurrencies by market cap.",
     "2.List all coins where circulating supply exceeds 90% of total supply.",
     "3. Get coins that are within 10% of their all-time-high (ATH).",
     "4. Find the average market cap rank of coins with volume above $1B.",
     "5. Get the most recently updated coin."
     )
)
     if option == "1. Find the top 3 cryptocurrencies by market cap.":
         filtered_df = pd.read_sql("SELECT name, symbol, market_cap FROM Cryptocurrencies ORDER BY market_cap DESC LIMIT 3",con=conn)
         st.dataframe(filtered_df)
     if option == "2.List all coins where circulating supply exceeds 90% of total supply.":
         filtered_df = pd.read_sql("SELECT name, symbol, circulating_supply, total_supply FROM Cryptocurrencies WHERE circulating_supply >= 0.9 * total_supply;",con=conn)
         st.dataframe(filtered_df)
     if option == "3. Get coins that are within 10% of their all-time-high (ATH).":
         filtered_df = pd.read_sql("SELECT name, symbol, current_price, ath FROM cryptocurrencies WHERE CAST(current_price AS REAL) >= 0.9 * CAST(ath AS REAL);",con=conn)
         st.dataframe(filtered_df)
     if option == "4. Find the average market cap rank of coins with volume above $1B.":
         filtered_df = pd.read_sql("SELECT AVG(market_cap_rank) AS avg_market_cap_rank FROM cryptocurrencies WHERE Date > 1000000000;",con=conn)
         st.dataframe(filtered_df)
     if option == "5. Get the most recently updated coin.":
         filtered_df = pd.read_sql("SELECT  * FROM cryptocurrencies ORDER BY Date DESC LIMIT 1;",con=conn)
         st.dataframe(filtered_df)

#crypto_prices
#st.title("Crypto Prices")
if selected == "Queries":
     option = st.selectbox(
     "crypto_prices",
    (
     "1. Find the highest daily price of Bitcoin in the last 365 days.",
     "2. Calculate the average daily price of Ethereum in the past 1 year.",
     "3. Show the daily price trend of Bitcoin in January 2025",
     "4. Find the coin with the highest average price over 1 year.",
     "5. Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025."
     )
)
     if option == "1. Find the highest daily price of Bitcoin in the last 365 days.":
         crypto_prices = pd.read_sql("SELECT MAX(price_usd) AS highest_price FROM crypto_price WHERE coin_id = 'bitcoin';",con=conn)
         st.dataframe(crypto_prices)
     if option == "2. Calculate the average daily price of Ethereum in the past 1 year.":
         crypto_prices = pd.read_sql("SELECT AVG(price_usd) AS avg_price FROM crypto_price WHERE coin_id = 'ethereum'",con=conn)
         st.dataframe(crypto_prices)
     if option == "3. Show the daily price trend of Bitcoin in January 2025":
         crypto_prices = pd.read_sql(" SELECT date, price_usd FROM crypto_price WHERE coin_id = 'bitcoin' and date between '2025-01-01' and '2025-04-30' ",con = conn)
         st.dataframe(crypto_prices)
     if option == "4. Find the coin with the highest average price over 1 year.":
         crypto_prices = pd.read_sql("SELECT coin_id, AVG(price_usd) AS avg_price FROM crypto_price GROUP BY coin_id ORDER BY avg_price DESC LIMIT 1;",con=conn)
         st.dataframe(crypto_prices)
     if option == "5. Get the % change in Bitcoin’s price between Sep 2024 and Sep 2025.":
         crypto_prices = pd.read_sql("SELECT ((sep2025.price - sep2024.price) * 100.0 / sep2024.price) AS pct_change FROM (SELECT price_usd AS price FROM crypto_price WHERE coin_id = 'bitcoin' AND date BETWEEN '2024-09-01' AND '2024-09-30' ORDER BY date DESC LIMIT 1) AS sep2024,(SELECT price_usd AS price FROM crypto_price WHERE coin_id = 'bitcoin' AND date BETWEEN '2025-09-01' AND '2025-09-30' ORDER BY date DESC LIMIT 1) AS sep2025;",con=conn)
         st.dataframe(crypto_prices)

#st.title("Oil prices")
if selected == "Queries":

     option = st.selectbox(
    "Oil_prices",
    (
     "1. Find the highest oil price in the last 5 years.",
     "2. Get the average oil price per year.",
     "3. Show oil prices during COVID crash (March–April 2020).",
     "4. Find the lowest price of oil in the last 10 years.",
     "5. Calculate the volatility of oil prices (max-min difference per year).  "
     )
)

     if option == "1. Find the highest oil price in the last 5 years.":
         filtered_df = pd.read_sql("SELECT MAX(price_usd) AS highest_price FROM oil_prices WHERE date >= date('now','-5 years');",con=conn)
         st.dataframe(filtered_df)
     if option == "2. Get the average oil price per year.":
         filtered_df = pd.read_sql("SELECT strftime('%Y', date) AS year, AVG(price_usd) AS avg_price FROM oil_prices GROUP BY year ORDER BY year;",con=conn)
         st.dataframe(filtered_df)
     if option == "3. Show oil prices during COVID crash (March–April 2020).":
         filtered_df = pd.read_sql("SELECT date, price_usd FROM oil_prices WHERE date BETWEEN '2020-03-01' AND '2020-04-30' GROUP BY date ORDER BY date;",con=conn)
         st.dataframe(filtered_df)
     if option == "4. Find the lowest price of oil in the last 10 years.":
         filtered_df = pd.read_sql("SELECT MIN(price_usd) FROM oil_prices WHERE date >= date('now','-10 years');",con=conn)
         st.dataframe(filtered_df)
     if option == "5. Calculate the volatility of oil prices (max-min difference per year).":
         filtered_df = pd.read_sql("SELECT strftime('%Y', date) AS year, MAX(price_usd) - MIN(price_usd) AS volatility FROM oil_prices GROUP BY year ORDER BY year;",con=conn)
         st.dataframe(filtered_df)

#"stock_prices"
#st.title("stock_prices")
if selected == "Queries":

     option = st.selectbox(
    "stock_prices",
    (
     "1. Get all stock prices for a given ticker",
     "2. Find the highest closing price for NASDAQ (^IXIC)",
     "3. List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)",
     "4. Get monthly average closing price for each ticker",
     "5. Get average trading volume of NSEI in 2024"
     )
)

     if option == "1. Get all stock prices for a given ticker":
         all_stocks = pd.read_sql("SELECT * FROM stock_prices WHERE ticker='^NSEI';",con=conn)
         st.dataframe(all_stocks)
     if option == "2. Find the highest closing price for NASDAQ (^IXIC)":
         all_stocks = pd.read_sql("SELECT MAX(close) AS highest_close FROM stock_prices WHERE ticker='^IXIC';",con=conn)
         st.dataframe(all_stocks)
     if option == "3. List top 5 days with highest price difference (high - low) for S&P 500 (^GSPC)":
         all_stocks = pd.read_sql("SELECT date, high, low, (high-low) AS price_diff FROM stock_prices WHERE ticker='^GSPC' ORDER BY price_diff DESC LIMIT 5;",con=conn)
         st.dataframe(all_stocks)
     if option == "4. Get monthly average closing price for each ticker":
         all_stocks = pd.read_sql("SELECT strftime('%Y-%m', date) AS month, ticker, AVG(close) AS avg_close FROM stock_prices GROUP BY ticker, month ORDER BY ticker, month;",con=conn)
         st.dataframe(all_stocks)
     if option == "5. Get average trading volume of NSEI in 2024":
         all_stocks = pd.read_sql("SELECT AVG(volume) AS avg_volume FROM stock_prices WHERE ticker='^NSEI' AND strftime('%Y', date)='2024';",con=conn)
         st.dataframe(all_stocks)

if selected == "Queries":
     option = st.selectbox(
     "Join Query",
    (
     "1. Compare Bitcoin vs Oil average price in 2025.",
     "2. Check if Bitcoin moves with S&P 500 (correlation idea).",
     "3. Compare Ethereum and NASDAQ daily prices for 2025.",
     "4. Find days when oil price spiked and compare with Bitcoin price change.",
     "5. Compare top 3 coins daily price trend vs Nifty (^NSEI).",
     "6. Compare stock prices (^GSPC) with crude oil prices on the same dates",
     "7. Correlate Bitcoin closing price with crude oil closing price (same date)",
     "8. Compare NASDAQ (^IXIC) with Ethereum price trends",
     "9. Join top 3 crypto coins with stock indices for 2025",
     "10. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison"
     )
)
     if option == "1. Compare Bitcoin vs Oil average price in 2025.":
         filtered_df = pd.read_sql("SELECT 'Bitcoin' AS Asset, AVG(price_usd) AS Avg_price_2025 FROM crypto_price WHERE strftime('%Y', date) = '2025' UNION ALL SELECT 'Oil' AS asset, AVG(price_usd) AS avg_price_2025 FROM oil_prices WHERE strftime('%Y', date) = '2025' ",con=conn)
         st.dataframe(filtered_df)
     if option == "2. Check if Bitcoin moves with S&P 500 (correlation idea).":
         filtered_df = pd.read_sql("SELECT s.date, s.close AS sp500_close, c.price_usd AS bitcoin_price FROM stock_prices s JOIN crypto_price c ON s.date = c.date WHERE s.ticker='^GSPC' AND c.coin_id='bitcoin' ORDER BY s.date",con=conn)
         st.dataframe(filtered_df)
     if option == "3. Compare Ethereum and NASDAQ daily prices for 2025.":
         filtered_df = pd.read_sql("SELECT s.date, s.close AS nasdaq_close, c.price_usd AS ethereum_price FROM stock_prices s JOIN crypto_price c ON s.date = c.date WHERE s.ticker='^IXIC' AND c.coin_id='ethereum' AND strftime('%Y', s.date)='2025' ORDER BY s.date",con=conn)
         st.dataframe(filtered_df)
     if option == "4. Find days when oil price spiked and compare with Bitcoin price change.":
         filtered_df = pd.read_sql(" SELECT o.date, o.price_usd AS oil_price, (o.price_usd - LAG(o.price_usd) OVER (ORDER BY o.date)) AS oil_change, c.price_usd AS bitcoin_price, (c.price_usd - LAG(c.price_usd) OVER (ORDER BY c.date)) AS bitcoin_change FROM oil_prices o LEFT JOIN crypto_price c ON o.date = c.date WHERE c.coin_id='bitcoin' ORDER BY o.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "5. Compare top 3 coins daily price trend vs Nifty (^NSEI).":
         filtered_df = pd.read_sql(" SELECT c.date, c.coin_id , c.price_usd AS crypto_price, s.close AS nifty_price FROM crypto_price c JOIN stock_prices s ON c.date = s.date WHERE c.coin_id IN ( SELECT coin_id FROM crypto_price GROUP BY coin_id ORDER BY AVG(price_usd) DESC LIMIT 3 ) AND s.ticker = '^NSEI' ORDER BY c.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "6. Compare stock prices (^GSPC) with crude oil prices on the same dates":
         filtered_df = pd.read_sql("SELECT s.date, s.close AS sp500_close, o.price_usd AS oil_price FROM stock_prices s LEFT JOIN oil_prices o ON s.date = o.date WHERE s.ticker='^GSPC' GROUP BY s.date ORDER BY s.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "7. Correlate Bitcoin closing price with crude oil closing price (same date)":
         filtered_df = pd.read_sql("SELECT c.date, c.price_usd AS bitcoin_price, o.price_usd AS oil_price FROM crypto_price c JOIN oil_prices o ON c.date = o.date WHERE c.coin_id='bitcoin' ORDER BY c.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "8. Compare NASDAQ (^IXIC) with Ethereum price trends":
         filtered_df = pd.read_sql("SELECT  s.date, s.close AS nasdaq_close, c.price_usd AS ethereum_price FROM stock_prices s JOIN crypto_price c ON s.date = c.date WHERE s.ticker='^IXIC' AND c.coin_id='ethereum' ORDER BY s.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "9. Join top 3 crypto coins with stock indices for 2025":
         filtered_df = pd.read_sql("SELECT c.date, c.coin_id, c.price_usd, s.ticker, s.close AS stock_close FROM crypto_price c JOIN stock_prices s ON c.date = s.date WHERE c.coin_id IN ('bitcoin','ethereum','tether') AND s.ticker IN ('^IXIC','^GSPC') AND strftime('%Y', c.date)='2025' ORDER BY c.date;",con=conn)
         st.dataframe(filtered_df)
     if option == "10. Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison":
         filtered_df = pd.read_sql("SELECT s.date, s.ticker, s.close AS stock_close, o.price_usd AS oil_price, c.price_usd AS bitcoin_price FROM stock_prices s JOIN oil_prices o ON s.date = o.date JOIN crypto_price c ON s.date = c.date WHERE s.ticker='^IXIC' AND c.coin_id='bitcoin' ORDER BY s.date;",con=conn)
         st.dataframe(filtered_df)

if selected == "Crypto Analysis":
    st.title("Cryptocurrency Price Dashboard")
# Select coin
    coin = st.selectbox(
    "Select Cryptocurrency",
    ["bitcoin","ethereum","tether"]
    )
    # Date filters
    start_date = st.date_input("Start Date")
    end_date = st.date_input("End Date")

# SQL Query
    query = """ SELECT date, price_usd FROM crypto_price WHERE coin_id = ? AND date BETWEEN ? AND ? ORDER BY date """
    df = pd.read_sql(query, conn, params=(coin, start_date, end_date))
# Show line chart
    st.subheader("Daily Price Trend")
    st.line_chart(df.set_index("date"))

# Show table
    st.subheader("Daily Price Table")
    st.dataframe(df)
```

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('ds.db')   # change the name of database accordingly
cursor = conn.cursor()
cursor.execute('''CREATE TABLE IF NOT EXISTS Crypto_price (
    coin_id VARCHAR(50),
    date DATE,
    price_usd DECIMAL(18, 6),

    FOREIGN KEY (coin_id)
    REFERENCES Cryptocurrencies(id)
    ON DELETE CASCADE
    ON UPDATE CASCADE
);
             ''')

conn.commit()
conn.close()
```
